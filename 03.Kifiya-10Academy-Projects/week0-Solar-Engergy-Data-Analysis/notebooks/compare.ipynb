```python
# üß© Section 1: Imports and Setup

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown as md
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
features = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB', 'Tamb']
```

```python
# üóÇÔ∏è Section 2: Load and Tag Datasets

# Load cleaned CSVs
benin = pd.read_csv('./data/benin-malanville_cleaned.csv')
sierraleone = pd.read_csv('./data/sierraleone-bumbuna_cleaned.csv')
togo = pd.read_csv('./data/togo-dapaong_qc_cleaned.csv')

# Add country labels
benin['Country'] = 'Benin-Malanville'
sierraleone['Country'] = 'SierraLeone-Bumbuna'
togo['Country'] = 'Togo-Dapaong'

# Combine all
df = pd.concat([benin, sierraleone, togo], ignore_index=True)
df.head()
```

```python
# üìä Section 3: Summary Statistics by Country

summary = df.groupby('Country')[features].describe().round(2)
summary
```

```python
# üì¶ Section 4: Boxplots of Each Feature by Country

def boxplot_feature(feature):
    plt.figure()
    sns.boxplot(x='Country', y=feature, data=df, palette="Set2")
    plt.title(f'{feature} Distribution by Country')
    plt.xticks(rotation=15)
    plt.tight_layout()
    plt.show()

# Loop over relevant features
for feat in features:
    boxplot_feature(feat)
```

```python
# üîÅ Section 5: Correlation Heatmap for Each Country

for country, data in df.groupby('Country'):
    plt.figure(figsize=(8, 6))
    sns.heatmap(data[features].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Correlation Matrix - {country}')
    plt.tight_layout()
    plt.show()
```
